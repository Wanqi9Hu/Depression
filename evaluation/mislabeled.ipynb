```python
import pandas as pd
import numpy as np
```

# #BERT

```python
bert_wrong_result=pd.read_csv('bert_wrong_result.csv')
bert_wrong_result=bert_wrong_result.dropna(axis=0,subset = ["normalized_text"])
bert_wrong_result.columns=['text','normalized_text','lable','result']
bert_wrong_result
```

```python
bert_wrong_result_list=[]
for i in range(0,bert_wrong_result.shape[0]):
    temp=bert_wrong_result.iloc[i,1].split()
    for n in range(0,len(temp)):
        bert_wrong_result_list.append(temp[n])
```

```python
from collections import Counter
wd = Counter(bert_wrong_result_list)
wd.most_common(200)
```

```python
bert_dic={}
for k,v in wd.items():
    bert_dic[k]=v
print(bert_dic['sorry'])
print(bert_dic['miss'])
print(bert_dic['mad'])
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wordcloud = WordCloud(
        background_color="white", 
        width=1500,              
        height=960,              
        margin=10               
        ).generate(" ".join(bert_wrong_result_list))

plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
